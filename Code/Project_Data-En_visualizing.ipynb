```python
# ... (‡∏™‡πà‡∏ß‡∏ô import library ‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏° DataFrame df_all ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏î‡∏¥‡∏°) ...
from collections import Counter
# Import the WordCloud class
from wordcloud import WordCloud

# --- 1. Word Cloud (‡∏õ‡∏£‡∏±‡∏ö‡∏™‡πà‡∏ß‡∏ô‡∏ô‡∏µ‡πâ) ---
# ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏à‡∏≤‡∏Å df_all ‡πÄ‡∏û‡∏∑‡πà‡∏≠ Word Cloud
product_types = df_all['Type'].tolist() # ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å column 'Type' ‡πÄ‡∏õ‡πá‡∏ô list
product_type_counts = Counter(product_types) # ‡∏ô‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤

wordcloud = WordCloud(width=800, height=400,
                      background_color='white').generate_from_frequencies(product_type_counts)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Product Type Word Cloud (Shopee & Lazada)', fontsize=14) # ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô Title ‡πÉ‡∏´‡πâ‡∏™‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏´‡∏°‡∏≤‡∏¢
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of Product Type Word Cloud (Shopee & Lazada)]
```

```python
# ... (‡∏™‡πà‡∏ß‡∏ô import library ‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏° DataFrame df_all ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏î‡∏¥‡∏°) ...

# --- 2. Bar Chart (‡∏õ‡∏£‡∏±‡∏ö‡∏™‡πà‡∏ß‡∏ô‡∏ô‡∏µ‡πâ) ---
# ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ï‡πà‡∏≠‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏à‡∏≤‡∏Å df_all
type_counts_bar = df_all['Type'].value_counts().nlargest(20) # ‡∏ô‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡πÅ‡∏•‡∏∞‡πÄ‡∏•‡∏∑‡∏≠‡∏Å Top 20 ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó
types_bar = type_counts_bar.index.tolist() # ‡∏î‡∏∂‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏à‡∏≤‡∏Å index
counts_bar = type_counts_bar.values.tolist() # ‡∏î‡∏∂‡∏á‡∏Ñ‡πà‡∏≤‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà (‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤)

plt.figure(figsize=(12, 6))
sns.barplot(x=counts_bar, y=types_bar, palette="viridis")
plt.xlabel('Number of Products') # ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô Label ‡πÅ‡∏Å‡∏ô X ‡πÉ‡∏´‡πâ‡∏™‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏´‡∏°‡∏≤‡∏¢
plt.ylabel('Product Type') # ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô Label ‡πÅ‡∏Å‡∏ô Y ‡πÉ‡∏´‡πâ‡∏™‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏´‡∏°‡∏≤‡∏¢
plt.title('Number of Products per Product Type (Top 20)', fontsize=14) # ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô Title
plt.gca().invert_yaxis()
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of Bar Chart of Number of Products per Product Type (Top 20)]
```

```python
# 3.1 ‡∏£‡∏≤‡∏Ñ‡∏≤‡∏Ç‡∏≤‡∏¢‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ (Bar Chart)
avg_price_per_type = df_all.groupby('Type')['Price'].mean().sort_values(ascending=False) # ‡∏£‡∏≤‡∏Ñ‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏ï‡πà‡∏≠‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤

plt.figure(figsize=(14, 7))
sns.barplot(x=avg_price_per_type.values, y=avg_price_per_type.index, palette='plasma') # ‡πÉ‡∏ä‡πâ palette 'plasma'
plt.xlabel('Average Price')
plt.ylabel('Product Type')
plt.title('‡∏£‡∏≤‡∏Ñ‡∏≤‡∏Ç‡∏≤‡∏¢‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏ï‡πà‡∏≠‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of Bar Chart of Average Price per Product Type]


# 3.1 ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏£‡∏≤‡∏Ñ‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÉ‡∏ô‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô (Box Plot)
plt.figure(figsize=(16, 8))
sns.boxplot(x='Type', y='Price', data=df_all, palette='Set2') # ‡πÉ‡∏ä‡πâ palette 'Set2'
plt.xlabel('Product Type')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right') # ‡∏õ‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏´‡∏°‡∏∏‡∏ô ‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏ß‡∏≤‡∏á tick labels ‡πÅ‡∏Å‡∏ô X
plt.tight_layout()
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of Box Plot of Price Distribution per Product Type]
```

```python
# 3.2 ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤‡∏Å‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢ (Scatter Plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Price', y='Quantity', data=df_all, alpha=0.5) # alpha=0.5 ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡∏à‡∏∏‡∏î‡πÇ‡∏õ‡∏£‡πà‡∏á‡πÅ‡∏™‡∏á ‡∏°‡∏≠‡∏á‡πÄ‡∏´‡πá‡∏ô‡∏à‡∏∏‡∏î‡∏ó‡∏µ‡πà‡∏ã‡πâ‡∏≠‡∏ô‡∏Å‡∏±‡∏ô
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.title('‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÅ‡∏•‡∏∞‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢')
plt.tight_layout()
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of Scatter Plot of Price vs Quantity Sold]
```

```python
# 4.1 ‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡πÉ‡∏î‡∏Ç‡∏≤‡∏¢‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î (Bar Chart - ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤)
best_selling_types_qty = df_all.groupby('Type')['Quantity'].sum().sort_values(ascending=False).head(10) # ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î 10 ‡∏≠‡∏±‡∏ô‡∏î‡∏±‡∏ö‡πÅ‡∏£‡∏Å (‡∏ß‡∏±‡∏î‡∏à‡∏≤‡∏Å‡∏à‡∏≥‡∏ô‡∏ß‡∏ô)

plt.figure(figsize=(14, 7))
sns.barplot(x=best_selling_types_qty.values, y=best_selling_types_qty.index, palette='rocket') # ‡πÉ‡∏ä‡πâ palette 'rocket'
plt.xlabel('Total Quantity Sold')
plt.ylabel('Product Type')
plt.title('‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î (‡∏ß‡∏±‡∏î‡∏à‡∏≤‡∏Å‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤) - Top 10')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of Bar Chart of Best Selling Product Types by Quantity (Top 10)]


# 4.2 ‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ‡∏£‡∏ß‡∏°‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ (Bar Chart - ‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ)
df_all['Revenue'] = df_all['Price'] * df_all['Quantity'] # ‡∏™‡∏£‡πâ‡∏≤‡∏á column 'Revenue' = Price * Quantity
revenue_per_type = df_all.groupby('Type')['Revenue'].sum().sort_values(ascending=False).head(10) # ‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ‡∏£‡∏ß‡∏°‡∏ï‡πà‡∏≠‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤

plt.figure(figsize=(14, 7))
sns.barplot(x=revenue_per_type.values, y=revenue_per_type.index, palette='mako') # ‡πÉ‡∏ä‡πâ palette 'mako'
plt.xlabel('Total Revenue')
plt.ylabel('Product Type')
plt.title('‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ‡∏£‡∏ß‡∏°‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ - Top 10')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of Bar Chart of Total Revenue per Product Type (Top 10)]
```

```python
# 4.3 ‡∏™‡∏±‡∏î‡∏™‡πà‡∏ß‡∏ô‡∏¢‡∏≠‡∏î‡∏Ç‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ (Pie Chart)
sales_proportion_type = df_all.groupby('Type')['Quantity'].sum().sort_values(ascending=False).head(10) # ‡∏¢‡∏≠‡∏î‡∏Ç‡∏≤‡∏¢‡∏£‡∏ß‡∏°‡∏ï‡πà‡∏≠‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ (Top 10)

plt.figure(figsize=(10, 10))
plt.pie(sales_proportion_type, labels=sales_proportion_type.index, autopct='%1.1f%%', startangle=140,
        colors=sns.color_palette('pastel')) # ‡πÉ‡∏ä‡πâ palette 'pastel' ‡πÅ‡∏•‡∏∞‡∏õ‡∏£‡∏±‡∏ö startangle
plt.title('Proportion sales of each product type (Top 10)')
plt.axis('equal')  # ‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏™‡πà‡∏ß‡∏ô‡∏†‡∏≤‡∏û‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô ‡∏ß‡∏á‡∏Å‡∏•‡∏°‡πÑ‡∏°‡πà‡πÄ‡∏ö‡∏µ‡πâ‡∏¢‡∏ß
plt.tight_layout()
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of Pie Chart of Sales Proportion per Product Type (Top 10)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate sales proportion
sales_proportion_type = df_all.groupby('Type')['Quantity'].sum().sort_values(ascending=False).head(10)

# Create figure with larger size and higher DPI for better quality
plt.figure(figsize=(12, 8), dpi=100)

# Create pie chart with improved styling
plt.pie(
    sales_proportion_type,
    labels=None,  # Remove labels from pie chart
    autopct='%.1f%%',
    startangle=90,
    counterclock=False,
    colors=sns.color_palette('husl', n_colors=10),
    wedgeprops={
        'edgecolor': 'white',
        'linewidth': 2,
        'antialiased': True
    },
    textprops={
        'fontsize': 12,
        'color': 'black'
    },
    pctdistance=0.85
)

# Make the chart circular
plt.axis('equal')

# Add legend
plt.legend(
    sales_proportion_type.index,
    title="Product Types",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

# Adjust layout to prevent text cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 5.2 ‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ô‡∏¥‡∏¢‡∏°‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡πÑ‡∏´‡∏ô (Bar Chart - Top Types per Province)
top_types_per_province = df_all.groupby(['Address', 'Type_Eng'])['Quantity'].sum().groupby(level=0).nlargest(5).reset_index(level=0, drop=True).reset_index()
# ‡∏´‡∏≤ Top 5 ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î

# ‡∏™‡∏£‡πâ‡∏≤‡∏á FacetGrid ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÅ‡∏™‡∏î‡∏á Bar Chart ‡πÅ‡∏¢‡∏Å‡∏ï‡∏≤‡∏°‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î
g = sns.FacetGrid(top_types_per_province, col='Province', col_wrap=4, height=4, aspect=1.2) # ‡πÅ‡∏ö‡πà‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏ï‡∏≤‡∏°‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î, 4 ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ï‡πà‡∏≠‡πÅ‡∏ñ‡∏ß
g.map(sns.barplot, 'Quantity', 'Type_Eng', palette='viridis') # ‡∏™‡∏£‡πâ‡∏≤‡∏á Bar Chart ‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞ Facet (‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î)

# ‡∏õ‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏á FacetGrid
g.set_titles(col_template="{col_name}") # ‡πÉ‡∏ä‡πâ‡∏ä‡∏∑‡πà‡∏≠‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î‡πÄ‡∏õ‡πá‡∏ô Title ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞ Facet
g.set_axis_labels('Total Quantity Sold', 'Product Type') # Label ‡πÅ‡∏Å‡∏ô‡∏£‡∏ß‡∏°
for ax in g.axes.flat: # ‡∏õ‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞ Axes (‡∏Å‡∏£‡∏≤‡∏ü‡∏¢‡πà‡∏≠‡∏¢)
    ax.set_xlabel('Total Quantity Sold') # Label ‡πÅ‡∏Å‡∏ô X ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Å‡∏£‡∏≤‡∏ü‡∏¢‡πà‡∏≠‡∏¢
    ax.set_ylabel('Product Type') # Label ‡πÅ‡∏Å‡∏ô Y ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Å‡∏£‡∏≤‡∏ü‡∏¢‡πà‡∏≠‡∏¢
    ax.invert_yaxis() # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á Bar Chart ‡∏à‡∏≤‡∏Å‡∏°‡∏≤‡∏Å‡πÑ‡∏õ‡∏ô‡πâ‡∏≠‡∏¢

g.fig.suptitle('‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏Ç‡∏≤‡∏¢‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î (Top 5)', y=1.05) # Title ‡πÉ‡∏´‡∏ç‡πà‡∏Ç‡∏≠‡∏á FacetGrid
plt.tight_layout()
plt.show()
# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏õ‡πá‡∏ô [Image of FacetGrid of Top 5 Product Types per Province]
```
