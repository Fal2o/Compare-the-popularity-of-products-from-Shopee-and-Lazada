```python
# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡πà‡∏≤‡∏Å‡∏≤‡∏£‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv("lazada_product.csv")
df
```

```python
# ‡πÅ‡∏¢‡∏Å column ‡πÉ‡∏´‡∏°‡πà‡∏à‡∏≤‡∏Å column "Product"
new_columns = df['Product'].str.split('‡∏ø', n=1, expand=True)
df['New_Column'] = new_columns[1]

df
```

```python
# ‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏ò‡∏≠‡∏î str.extract() ‡∏£‡πà‡∏ß‡∏°‡∏Å‡∏±‡∏ö regular expression ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏±‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£
df['Price'] = df['New_Column'].str.extract(r'(\d+\.\d{2})')

# ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ "‡∏ö‡∏≤‡∏ó" ‡∏ï‡πà‡∏≠‡∏ó‡πâ‡∏≤‡∏¢‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÇ‡∏î‡∏¢‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á
# df['Price'] = df['Price'].str.replace(r'(\d+\.\d{2})', r'\1 ‡∏ö‡∏≤‡∏ó', regex=True)
df
```

```python
# ‡πÉ‡∏ä‡πâ‡πÄ‡∏°‡∏ò‡∏≠‡∏î str.extract() ‡∏£‡πà‡∏ß‡∏°‡∏Å‡∏±‡∏ö regular expression ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏±‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£
df['Quantity'] = df['New_Column'].str.extract(r'(‡∏ä‡∏¥‡πâ‡∏ô.*?\))')

# ‡∏•‡∏ö‡∏ß‡∏á‡πÄ‡∏•‡πá‡∏ö‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏≠‡∏≠‡∏Å‡∏à‡∏≤‡∏Å column 'Quantity'
df['Quantity'] = df['Quantity'].str.replace(r'\(|\)', ' ', regex=True)

# ‡∏•‡∏ö‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ "‡∏ä‡∏¥‡πâ‡∏ô" ‡∏≠‡∏≠‡∏Å‡∏à‡∏≤‡∏Å column 'Quantity'
df['Quantity'] = df['Quantity'].str.replace('‡∏ä‡∏¥‡πâ‡∏ô', '')

df
```

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÅ‡∏¢‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏´‡∏°‡πà
def extract_address(row):
    parts = row['New_Column'].split(')')
    if len(parts) > 1:
        return parts[1]
    else:
        parts = row['New_Column'].rsplit('f', 1)
        return parts[-1] if len(parts) > 1 else ''

# ‡πÉ‡∏ä‡πâ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á column 'Address'
df['Address'] = df.apply(extract_address, axis=1)

df
```

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏•‡∏ö‡∏Ñ‡∏≥‡πÉ‡∏ô column 'Product' ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ö‡∏Ñ‡∏≥‡πÉ‡∏ô column 'New_Column'
def remove_words_from_product(row):
    product = row['Product']
    for word in row['New_Column'].split():
        product = product.replace(word, '')
    return product

# ‡πÉ‡∏ä‡πâ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏•‡∏ö‡∏Ñ‡∏≥‡πÉ‡∏ô column 'Product'
df['Product'] = df.apply(remove_words_from_product, axis=1)

# ‡∏•‡∏ö‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£ "‡∏ø" ‡∏ó‡∏µ‡πà‡∏ï‡πà‡∏≠‡∏ó‡πâ‡∏≤‡∏¢‡∏≠‡∏¢‡∏π‡πà‡∏≠‡∏≠‡∏Å‡∏à‡∏≤‡∏Å column 'Product'
df['Product'] = df['Product'].str.replace('‡∏ø', '')

df
```
